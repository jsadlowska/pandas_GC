```python
import pandas as pd
import numpy as np
```

```python
f_results = pd.read_excel('data/female_concentrations.xlsx')
m_results = pd.read_excel('data/male_concentrations_after_q_dixon.xlsx')
m_concentrations = pd.read_excel('data/male_concentrations.xlsx')
```

```python
## To obtain the final results, it is necessary to determine the arithmetic mean, standard deviation, and measurement uncertainty.
```

```python
for col in f_results.columns[2:]:
    f_results[col] = pd.to_numeric(f_results[col], errors='coerce')

f_results['Concentration (µg/g)'] = f_results.iloc[:, 2:].mean(axis=1).round(2)
f_results['Standard Deviation'] = f_results.iloc[:, 2:].std(axis=1).round(2)

n_measurements = f_results.iloc[:, 2:].apply(lambda x: x.count(), axis=1) 
f_results['Measurement Uncertainty'] = (f_results['Standard Deviation'] / np.sqrt(n_measurements)).round(2)

f_results = f_results[['Name of the chemical compound', 'Concentration (µg/g)', 'Standard Deviation', 'Measurement Uncertainty']]

f_results
```

```python
f_results.to_excel('data/female_results.xlsx')
```

```python
m_results = m_concentrations.copy()

for col in m_results.columns[2:]:
    m_results[col] = pd.to_numeric(m_results[col], errors='coerce')

m_results['Concentration (µg/g)'] = m_results.iloc[:, 2:].mean(axis=1).round(2)
m_results['Standard Deviation'] = m_results.iloc[:, 2:].std(axis=1).round(2)

n_measurements = m_results.iloc[:, 2:].apply(lambda x: x.count(), axis=1) 
m_results['Measurement Uncertainty'] = (m_results['Standard Deviation'] / np.sqrt(n_measurements)).round(2)

m_results = m_results[['Name of the chemical compound', 'Concentration (µg/g)', 'Standard Deviation', 'Measurement Uncertainty']]

m_results
```

```python
m_results.to_excel('data/male_results.xlsx')
```
