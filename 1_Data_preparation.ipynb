```python
import pandas as pd
```

```python
female_df = pd.read_excel('data/female_data.xlsx', header = None)
male_df = pd.read_excel('data/male_data.xlsx', header = None)
```

```python
female_df
```

```python
male_df
```

```python
## To determine the concentrations of the given compounds, we need the peak area values for each compound in the analysis.
```

```python
female_peaks = female_df.drop([0, 1])
female_peaks = female_peaks.reset_index(drop=True)
female_peaks.columns = female_peaks.iloc[0]
female_peaks = female_peaks.drop(0).reset_index(drop=True)

female_peaks.columns = female_peaks.columns.astype(str)
female_peaks = female_peaks.drop(columns=[col for col in female_peaks.columns if 'time' in col or 'nan' in col])

new_names = ['FEMALE 1', 'FEMALE 2', 'FEMALE 3']

columns = female_peaks.columns

updated_columns = []
counter = 0

for col in columns:
    if 'peak area' in col:
        updated_columns.append(new_names[counter])
        counter += 1
    else:
        updated_columns.append(col)

female_peaks.columns = updated_columns
```

```python
female_peaks
```

```python
male_peaks = male_df.drop([0, 1])
male_peaks = male_peaks.reset_index(drop=True)
male_peaks.columns = male_peaks.iloc[0]
male_peaks = male_peaks.drop(0).reset_index(drop=True)

male_peaks.columns = male_peaks.columns.astype(str)
male_peaks = male_peaks.drop(columns=[col for col in male_peaks.columns if 'time' in col or 'nan' in col])

new_names = ['MALE 1', 'MALE 2', 'MALE 3', 'MALE 4']

columns = male_peaks.columns

updated_columns = []
counter = 0

for col in columns:
    if 'peak area' in col:
        updated_columns.append(new_names[counter])
        counter += 1
    else:
        updated_columns.append(col)

male_peaks.columns = updated_columns
```

```python
male_peaks
```

```python
female_peaks.to_excel('data/female_peaks.xlsx')
male_peaks.to_excel('data/male_peaks.xlsx')
```

```python
## To determine the concentrations of the given compounds, we also need the masses of the individual spiders.
```

```python
female_weights = female_df.drop(index=0).reset_index(drop=True)

female_weights.columns = female_weights.iloc[0]

female_weights = female_weights.drop(index=1).reset_index(drop=True)

female_weights = female_weights.iloc[:1]

female_weights = female_weights.dropna(axis=1, how='all')

new_names = ['FEMALE 1', 'FEMALE 2', 'FEMALE 3']
female_weights.columns = [None] + new_names 

female_weights.columns = [col if col != 'Weight of the spider in grams' else '' for col in female_weights.columns]

female_weights
```

```python
male_weights = male_df.drop(index=0).reset_index(drop=True)

male_weights.columns = male_weights.iloc[0]

male_weights = male_weights.drop(index=1).reset_index(drop=True)

male_weights = male_weights.iloc[:1]

male_weights = male_weights.dropna(axis=1, how='all')

new_names = ['MALE 1', 'MALE 2', 'MALE 3', 'MALE 4']
male_weights.columns = [None] + new_names 

male_weights.columns = [col if col != 'Weight of the spider in grams' else '' for col in male_weights.columns]

male_weights
```

```python
female_weights.to_excel('data/female_weights.xlsx')
male_weights.to_excel('data/male_weights.xlsx')
```
